```python
employees = [
{"name": "Alice", "completed_training": True},
{"name": "Bob", "completed_training": False},
{"name": "Charlie", "completed_training": True},
{"name": "Dana", "completed_training": False}
]
completed_training = []
for employee in employees:
 if employee["completed_training"]:
   completed_training.append(employee["name"]) # Add names of employees who completed training,".append" is a special function to manipulate the string.
print("Employees who completed the training:")
for name in completed_training:
  print(name)
```
