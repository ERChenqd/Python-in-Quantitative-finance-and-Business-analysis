```python
def age_calculation(birth_year, current_year):
    age = current_year - birth_year
    return age

birth_year = int(input("Enter your birth year: "))
current_year = int(input("Enter current year: "))
userage=age_calculation(birth_year,current_year)
print(userage)
```
