```python
#subtask 2
#Program 2
Employees_salary = {
'Tom':1800,
'Sam':1600,
'Sally':2200,
'Alex':2500,
'Rose':2100
}
# write a for loop to access each element of dictionary using its key
# use if statement to check the value of the key is less than 2000£
# if so, add 200£ to the value and store it as new value for that key
# you can use the dictionary method item() to print the final output
for key,value in Employees_salary.items():
    if value < 2000:
        Employees_salary[key] = value + 200
print(Employees_salary.items())
```

```python
#subtask 3
#Program 3
Employees_salary = {
'Tom':1800,
'Sam':1600,
'Sally':2200,
'Alex':2500,
'Rose':2100
}
# Hint: create an empty list first as below and append items based on logic
#inside loop and conditional check.
high_salary_employees = []
for employeename in Employees_salary.keys():
    if Employees_salary[employeename] > 2000:
        high_salary_employees.append(employeename)
print(high_salary_employees)
```
