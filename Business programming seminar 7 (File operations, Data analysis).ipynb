```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/IMDBMovies.csv')

# Inspect the first few rows of the "Gross" column to identify issues
print(df["Gross"].head())

# Preprocess the "Gross" column
df["Gross"] = df["Gross"].str.replace(',', '', regex=True)  # Remove commas
df["Gross"] = df["Gross"].str.strip()  # Remove leading/trailing spaces

# Convert 'Gross' to numeric, handling errors
df["Gross"] = pd.to_numeric(df["Gross"], errors='coerce')

# Check for non-NaN values and find the highest-grossing movie
if df["Gross"].notna().sum() > 0:  # Ensure valid data exists
    highest_gross_movie = df.loc[df["Gross"].idxmax()]
    print("Highest grossing movie:\n", highest_gross_movie[["Series_Title", "Gross"]])
else:
    print("No valid 'Gross' values found in the dataset.")
```
