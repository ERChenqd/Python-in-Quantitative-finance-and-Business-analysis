```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Example data: composition of attendances in 5 reasons and weekday pairs
data = {
    'Time': ['23:00', '23:00', '23:00', '23:00', '23:00',
             '1:00', '1:00', '1:00', '1:00', '1:00'],
    'Reason': ['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Reason5',
               'Reason1', 'Reason2', 'Reason3', 'Reason4', 'Reason5'],
    'Attendances': [50, 30, 20, 10, 15,   # Data at 23:00
                    60, 40, 25, 15, 20], # Data at 1:00
    'Weekday Pair': ['Sunday-Monday', 'Sunday-Monday', 'Sunday-Monday',
                     'Sunday-Monday', 'Sunday-Monday',
                     'Sunday-Monday', 'Sunday-Monday', 'Sunday-Monday',
                     'Sunday-Monday', 'Sunday-Monday']
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Pivot data for stacked bar plot
bar_data = df.pivot_table(index='Time', columns='Reason', values='Attendances', aggfunc='sum')

# Line data (attendances change across two time points)
line_data = df.groupby(['Weekday Pair', 'Time'])['Attendances'].sum().reset_index()

# Initialize a Matplotlib figure
fig, ax1 = plt.subplots(figsize=(10, 6))

# --- Plot stacked bar chart ---
bar_data.plot(kind='bar', stacked=True, ax=ax1, width=0.6, colormap='viridis', edgecolor='black')

# Labeling the Y-axis and X-axis for the bars
ax1.set_ylabel('Number of Attendances')
ax1.set_xlabel('Time')
ax1.set_title('Attendances Composition by Reason and Change Across Time')

# --- Plot lines for attendances change ---
# We use a secondary y-axis to overlay lines showing changes
ax2 = ax1.twinx()
for pair in line_data['Weekday Pair'].unique():
    subset = line_data[line_data['Weekday Pair'] == pair]
    ax2.plot(subset['Time'], subset['Attendances'], marker='o', linestyle='-', label=pair)

# Add legends
ax1.legend(title='Reasons', bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.legend(title='Weekday Pairs', bbox_to_anchor=(1.05, 0.5), loc='upper left')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
