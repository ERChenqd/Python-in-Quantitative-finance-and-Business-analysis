```python
customers_prefer=['Jing','Fang','Smriti','Samrat','Dave','Oliver','Sam']

def loyalty_points(name, loyalty_type):
    # Check if the customer is preferred
    if name not in customers_prefer:
        return None, "Sorry, this program is for preferred customers only."

    # Determine points required based on loyalty type
    if loyalty_type == "Platinum":
        points_needed = None
    elif loyalty_type == "Gold":
        points_needed = 5
    elif loyalty_type == "Silver":
        points_needed = 10
    elif loyalty_type == "None":
        points_needed = 0
    else:
        return None, "Invalid loyalty type."

    return points_needed, None

name = input("Can you type your name: ")
loyalty_type = input("Can you input your loyalty grade (Platinum, Gold, Silver, None): ")
# Call the function
points_needed, error_message = loyalty_points(name, loyalty_type)

# Output the result
if error_message:
    print(error_message)
else:
    print(f"Hey {name}, welcome to our site, hope you are having a good day")
    print(f"{name}, you qualify for a 15% discount on all our products, enjoy!")

    if points_needed is None:
        print("CALL CUSTOMER CARE FOR MEMBERSHIP TO AVAIL DISCOUNTS ON OUR PRODUCTS")
    elif points_needed == 5:
        print("You need 5 loyalty points more to become our Platinum member")
    elif points_needed == 10:
        print("You need 10 loyalty points more to become our Platinum member")
    elif points_needed == 0:
        print("CALL CUSTOMER CARE FOR MEMBERSHIP TO AVAIL DISCOUNTS ON OUR PRODUCTS")
```
