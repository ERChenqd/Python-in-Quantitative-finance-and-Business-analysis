```python
#Exercise 3
#task 1
purchase=int(input("Amount:"))
if purchase<=1000:
  disrate=0.05
  discount=purchase*disrate
  print("Discount:",discount)
elif purchase>1000 and purchase<=1500:
  disrate=0.1
  discount=purchase*disrate
  print("Discount:",discount)
else:
  disrate=0.15
  discount=purchase*disrate
  print("Discount:",discount)
print("Net payment: $",purchase-discount)
```
