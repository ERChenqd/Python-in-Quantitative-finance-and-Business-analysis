```python
customer_list = ["erwen", "linda"]
attempts = 3  # Number of attempts allowed

while attempts > 0:
    Vcode = input("What is your code? ")

    # Check if the input (converted to lowercase) is in the customer list
    if Vcode.lower() in customer_list:
        print(f"Welcome to the party, {Vcode}!")
        break  # Exit the loop if the code is correct
    else:
        attempts -= 1
        print(f"Sorry, you are not invited. You have {attempts} attempts left.")

# The else block runs only if the loop wasn't exited with a break
else:
    print("You have used all your attempts. Access denied.")
```
