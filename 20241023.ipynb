```python
import time
weight=float(input("Enter the weight in kg"))

start=time.time()

if weight<5:
  charge=5
elif weight<10:
  charge=10
elif weight<20:
  charge=15
else:
  charge=20

print("Service price:$",charge,"kg in package",weight)
end=time.time()
print("Time taken:",end-start)
# you must have an elif under if condition, you can't have an elif under else
```
