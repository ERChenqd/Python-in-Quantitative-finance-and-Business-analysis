```python
#Exercise A

# The plotting functionality of matplotlib is found in the pyplot
#submodule.
# Conventionally this is imported under the name plt.
import matplotlib.pyplot as plt
# Define two series of data to plot against each other.
x_points = range(0, 21) # generate elements from 0 to 20 (20 is the 21st element)
y_points = [0.3, 0.6, 0.8, 0.9, 1.3, 0.5, 0.2, 0.3, 0.2, 0.3, 0.7, 1.8,2.3, 1.7, 1.3, 2.6, 1.5, 1.4, 1.2, 0.5, 0.3]
# The plt.plot() function can be used to plot a simple line graph.
# The first two parameters are lists of the x and y coordinates of
# data points we wish to plot.
plt.plot(x_points, y_points)
# The labels of the graph can be set with the plt.title(),plt.xlabel(),
# and plt.ylabel() functions.
plt.title("A Fabulous Line Graph")
plt.xlabel("Abscissa")
plt.ylabel("Ordinate")
# When working in a Python notebook we can use the ply.show()
# function to show our plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Define two series of data to plot against each other.
x_points = range(0, 21)
y_points = [0.3, 0.6, 0.8, 0.9, 1.3, 0.5, 0.2, 0.3, 0.2, 0.3, 0.7, 1.8,
2.3, 1.7, 1.3, 2.6, 1.5, 1.4, 1.2, 0.5, 0.3]
# Plot a red line graph of our data by specifying "r" as the format
#string.
plt.plot(x_points, y_points, color="b")
plt.title("A Fabulous Line Graph")
plt.xlabel("Abscissa")
plt.ylabel("Ordinate")
plt.show()
```

```python
# Plot a magenta dashed line with x markers by specifying "x--m" as the
#format string.
plt.plot(x_points, y_points,"o--m",markerfacecolor="g")
plt.title("A Fabulous Line Graph")
plt.xlabel("Abscissa")
plt.ylabel("Ordinate")
# Add a grid to aid readability
plt.grid()
plt.show()
```

```python
#Task 2
import matplotlib.pyplot as plt

# Define two series of data to plot against each other.
y_points = range(0, 21)
x_points_1 = [4, 6, 7, 3, 5, 2, 7, 4, 8, 5, 6, 7, 4, 6, 5, 7, 5, 6, 8,
3, 1]
x_points_2 = [5, 3, 7, 4, 8, 9, 7, 8, 9, 6, 4, 2, 1, 3, 5, 3, 6, 4, 7,
3, 8]
# Plot each set of X values individually. The colour of each line
# will be automatically selected by matplotlib. The string passed to
# the label argument will be used in the legend to identify the data.
plt.plot(y_points, x_points_1, label="Data 1")
plt.plot(y_points, x_points_2, label="Data 2")
# Set title and axis labels.
plt.title("Two sets of Data on One Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
# Add a legend to the plot using the labels we defined previously.
#You can also position this label at various place on graph
plt.legend()
# Show the plot
plt.show()
```

```python
# prompt: plot all the X values in the X axis in the above graph

import matplotlib.pyplot as plt

# Define two series of data to plot against each other.
y_points = range(0, 21)
x_points_1 = [4, 6, 7, 3, 5, 2, 7, 4, 8, 5, 6, 7, 4, 6, 5, 7, 5, 6, 8,
              3, 1]
x_points_2 = [5, 3, 7, 4, 8, 9, 7, 8, 9, 6, 4, 2, 1, 3, 5, 3, 6, 4, 7,
              3, 8]

# Plot each set of X values individually.
plt.plot(y_points, x_points_1, label="Data 1", marker='o', linestyle='-') # Added markers and linestyle
plt.plot(y_points, x_points_2, label="Data 2", marker='x', linestyle='--') # Added markers and linestyle


# Set title and axis labels.
plt.title("Two sets of Data on One Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```

```python
#Bar chart:
# X axis labels, and list of bar heights.
labels = ["Cats", "Dogs", "Ants", "Camels", "Other"]
counts = [10, 15, 1, 2, 7]
# Plot the data as a bar chart.
plt.bar(labels, counts)
# Set axis labels.
plt.xlabel("Pets")
plt.ylabel("Count")
plt.show()
```

```python
#Count of observations of categorical variables/dummy variables
import seaborn as sns
import matplotlib.pyplot as plt
#predefined dataset called Titanic
df_titanic = sns.load_dataset("titanic")
df_titanic = df_titanic.rename(
columns={'sex':'Sex',
'class':'Class',
'alive':'Alive'})
# Plot countplot, using seaborn countplot function to plot the graph
ax = sns.countplot(
x="Class",
data=df_titanic,
hue='Alive')
# Set title - OPTIONAL , if you would want to further refine the grapgh with title
plt.title("Countplot example with titanic dataset")
plt.show()
```
