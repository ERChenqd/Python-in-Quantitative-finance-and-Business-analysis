```python
import pandas as pd
file_path="/content/faulty_data.csv"

df=pd.read_csv(file_path)
print(df)
```

```python
# Group by Department and calculate the average salary
grouped_df = df.groupby('Department')['Salary'].mean().reset_index()
grouped_df.rename(columns={'Salary': 'Average Salary'}, inplace=True)
print("Grouped Data by Department with Average Salary:")
print(grouped_df)
```
