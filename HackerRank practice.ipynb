```python
###Repeated String: find the number of 'a' in a repeated string, n is the length of the string, s is the repeated term of the string
def repeatedString(s, n):
    a_no=0
    s.split()
    for element in s:
        if element== "a":
            a_no +=1

    quotinent=n//len(s)
    a_no=quotinent*a_no
    modulo=n%len(s)
    remain=s[:modulo]

    for element in remain:
        if element== "a":
            a_no +=1

    return a_no
```

```python
###Matrix Calculation
def hourglassSum(arr):
    # want to find the maximum hourglass sum
    # minimum hourglass sum = -9 * 7 = -63
    max_sum = -63
    for i in range(4):
        for j in range(4):

            # sum of top 3 elements
            top = sum(arr[i][j:j+3])  #index[A:B] would only show the data starting from A to B-1

            # sum of the mid element
            mid = arr[i+1][j+1]

            # sum of bottom 3 elements
            bottom = sum(arr[i+2][j:j+3])

            hourglass = top + mid + bottom

            if hourglass > max_sum:
                max_sum = hourglass

    return max_sum
```

```python
###Left rotation: rotate the a[0] item to a[n-1] item (last term) for consecutively d times.
def rotLeft(a, d, n):
    rot=0
    d = d % n # find the effective number of rotations

    while rot<d:
        first_element= a[0]
        a.pop(0) #.pop() is more effective than .remove() for the first element
        a.append(first_element)
        rot+=1

    return a
##
##.pop(index)
#Purpose: Removes and returns an item at a specific index.
#Arguments: Takes an optional index (default is -1, meaning the last element).
#Behavior:
#If no index is given, removes the last element.
#Returns the removed element.
#Raises IndexError if the index is out of range.

#.remove(value)
#Purpose: Removes the first occurrence of a given value.
#Arguments: Takes a value (not an index).
#Behavior:
#Does not return the removed element.
#Raises ValueError if the value is not found.
#Time Complexity: O(n) (must search the entire list in worst case).
```

```python
###这个函数 minimumSwaps(n, arr) 的作用是计算将一个无序数组 arr 通过交换相邻元素的方式，计算升序排列所需的最少交换次数。
def minimumSwaps(n,arr):
    swap = 0
    visited = [False] * n
    #The visited array is initialized to False for all indices, indicating that no elements have been processed yet.
    #生成一个列表 arr_pos，其中每个元素是一个元组 (value, original_index)，存储原数组的值和它的原始位置
    #如果 arr = [4, 3, 2, 1]，则 arr_pos = [(4,0), (3,1), (2,2), (1,3)]
    arr_pos = [(arr[i], i) for i in range(n)]
    #对 arr_pos 按照 value 进行升序排序。排序后：arr_pos = [(1,3), (2,2), (3,1), (4,0)]（因为 1 < 2 < 3 < 4）
    arr_pos.sort()

    for i in range(n):
        # If already visited or in the correct position
        if visited[i] or arr_pos[i][1] == i:
            #遍历每个位置 i：如果 visited[i] 是 True，说明这个位置已经被处理过，跳过。如果 arr_pos[i][1] == i，说明这个位置的元素已经在正确的位置（排序后的位置），无需交换，跳过。
            continue
        # Start a new cycle
        # j = i：从当前位置 i 开始，沿着“应该放在这个位置的元素”的路径遍历。
        # visited[j] = True：标记当前元素为已访问。
        # j = arr_pos[j][1]：找到当前元素在原始数组中的位置，继续追踪。
        # cycle_size += 1：环的大小增加 1。
        cycle_size = 0
        j = i
        while not visited[j]:
            visited[j] = True
            j = arr_pos[j][1]
            cycle_size += 1
        # 例子（arr = [4, 3, 2, 1]）：
        # 初始 i = 0，arr_pos[0] = (1,3)，说明 1 应该在 0 的位置，但当前在 3 的位置。
        # j = 0 → visited[0] = True → j = arr_pos[0][1] = 3（跳到 3）。
        # visited[3] = True → j = arr_pos[3][1] = 0（回到 0，环闭合）。
        # cycle_size = 2（环 0 → 3 → 0）
        # Add the swaps for this cycle
        if cycle_size > 0:
            swap += (cycle_size - 1)

    return swap
# visited 数组的作用：避免重复处理已经归位的元素，确保每个环只计算一次。
# 环的检测：通过追踪“当前元素应该在的位置”来找到环。
# 交换次数的计算：每个环需要 (环大小 - 1) 次交换。
```

```python
###Array manipulation:对不定长度的array的任意区间[a,b]内进行增加k值的操作。array内elements初始值皆为0，操作完成后查找最大值。
# 源代码:对不定长度的array的任意区间[a,b]进行增加操作。array内elements初始值皆为0
def arrayManipulation(n,queries):
    arr=[0] * n
    for item in range(len(queries)):
        a=queries[item][0]
        b=queries[item][1]
        k=queries[item][2]
        for subject in range(a-1,b):
            arr[subject]=k + arr[subject]

    return max(arr)
###问题：执行时间过长，时间复杂度为O(n*m)
```

```python
###代码优化: 时间复杂度为O(n+m)
# 优化思路：差分数组（Difference Array）它可以在 O(1) 时间内完成区间增量操作，最后再通过前缀和计算原数组的值，这样总的时间复杂度降为 O(n + m),
# 差分数组（Difference Array）是一种高效处理 区间增量操作 的技术。它的核心思想是：不直接修改原数组，而是通过 标记区间的起点和终点，最后再通过 前缀和计算 还原原数组的最终状态。这样可以将区间操作的复杂度从 O(n) 降到 O(1)。
def arrayManipulation(n,queries):
    arr = [0] * (n + 2)  # Initialize an array with extra space to avoid index issues

    for a, b, k in queries:
        arr[a] += k
        arr[b + 1] -= k #不再逐个元素修改，而是通过 arr[a] += k 和 arr[b+1] -= k 标记区间。

    max_val = current = 0
    for i in range(1, n + 1):
        current += arr[i]
        if current > max_val:
            max_val = current

    return max_val
```

```python
###题目要求找出字符串中所有无序的变形子串对（anagrammatic pairs）的数量。
# 需要找到所有满足条件的子串对，这两个子串是变形词（即字符相同但顺序不同），且这两个子串在字符串中的位置不同。
# 关键洞察：对于任意两个子串，如果它们排序后相同，那么它们就是变形词
# 算法选择：
# 生成所有可能的子串。
# 对每个子串进行排序，以便比较是否为变形词。
# 使用字典来统计每个排序后的子串出现的次数。
# 对于每个出现次数大于1的排序子串，计算其组合数（即C(n,2)= nC2 = n*(n-1)//2）。
def sherlockAndAnagrams(s):
    substr_dict = {}
    # 生成所有可能的子串
    for i in range(len(s)):
        for j in range(i + 1, len(s) + 1):
            # 获取子串并排序
            substr = s[i:j] #substr：原始的子字符串，比如 "abc"、"cba"、"bac" 等
            sorted_substr = ''.join(sorted(substr)) #将子串转换为排序后的字符串形式
            # sorted(substr)：将字符串转换为字符列表并排序
            #''.join(sorted(substr))：将排序后的字符列表重新连接成字符串，例如：''.join(['a', 'b', 'c']) → "abc"
            # 统计排序后的子串出现次数
            substr_dict[sorted_substr] = substr_dict.get(sorted_substr, 0) + 1
            #substr_dict.get(sorted_substr, 0)：从字典中获取 sorted_substr 对应的值。如果 sorted_substr 不存在于字典中，返回默认值 0，如果存在，则返回 1。

    # 计算所有变形子串对的数量
    count = 0
    for key in substr_dict:
        n_val = substr_dict[key]
        if n_val >= 2:
            count += n_val * (n_val - 1) // 2

    return count
```
