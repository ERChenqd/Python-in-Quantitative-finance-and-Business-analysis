```python
s="I would do anything for you rather than be a cold fish"
```

```python
s[0] # indexing at position 0, which is the first character
```

```python
s[10] # indexing at position 10, which is the 10th character, a space
```

```python
s[:5] #slicing, we specify only the stop position, which content starts at the 0 position and ends before the 6th character
```

```python
s[5:] #slicing, we specify only the start position, which content starts after the 5th position and ends in the last character
```

```python
s[2:3] # slicing, target the content before the 2nd character and after the 4th character
```

```python
s[0:6:3] #inclusively between "I woul", the 1st character is "I", the 4(1+3)th character is "o"
```

```python
s[:] #copy the entire s
```
